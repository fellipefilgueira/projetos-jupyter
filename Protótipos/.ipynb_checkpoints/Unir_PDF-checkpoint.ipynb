```python
from PyPDF2 import PdfFileMerger
import pandas as pd
import os
import shutil
```

```python
caminho = r'D:/Users/CSR342486/IBERDROLA S.A/Cosern_Perdas - Documentos/TOI - Cálculo e Comunicação/Comunicação/COMUNICAÇÃO ABRIL 2022/Terceira_remessa'
camimho_arq = r'D:/Users/CSR342486/IBERDROLA S.A/Cosern_Perdas - Documentos/TOI - Cálculo e Comunicação/Comunicação/'
arq = [] 
number = []
for (root, dirs, file) in os.walk(caminho):    
    for f in file: 
        if '.pdf' in f: 
            arq.append(f)
for i in arq:
    number.append(str(i).replace('.pdf',"").replace('(0)',"").replace('(1)',"").replace('(2)',""))

n_dupl = dict.fromkeys(number).keys()
notas = [item for item in n_dupl]
display(notas)
```

```python
erros = []
num =[]
for k in notas: 
    pdfs = ['/'+str(k)+'(0).pdf','/'+str(k)+'(1).pdf','/'+str(k)+'(2).pdf']
    num.append(k)
    try:
    
        novo_arq = camimho_arq+str(k)+'.pdf'

        merger = PdfFileMerger()
        try:
            merger.append(caminho+pdfs[0], pages=(2,3))
        except Exception as erro:
            merger.append(caminho+pdfs[0])

        merger.append(caminho+pdfs[1])

        merger.append(caminho+pdfs[2])

        merger.write(novo_arq)

        merger.close()

        os.rename(novo_arq, caminho+'/Doc_Juntos/'+str(k)+'.pdf')

        
        erros.append("Ok")
        
    except OSError:
        erros.append("Erro")
        
display(erros)
df = pd.DataFrame(list(zip(num,erros)), columns = ['Notas','Verificação'])
df.to_excel(caminho+'/Doc_Juntos/'+'Doc_Juntos.xlsx')
```
